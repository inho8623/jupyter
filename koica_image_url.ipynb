```python
########### 최종코드 ###########
#  https://victorydntmd.tistory.com/245 // bs 관련 참고.

import requests
from bs4 import BeautifulSoup

import csv

print("*********************** 시작 ***********************")

last_page = 49 # 49 마지막 페이지

f = open('image_list.csv', 'w', newline='')
wr = csv.writer(f)
wr.writerow(['게시날짜','게시물명','링크'])

for page_num in range(1,last_page+1):
#'''
    print(page_num,"페이지 시작")
    page_url = 'http://www.koica.go.kr/bbs/koica_kr/147/artclList.do?page='+str(page_num)
    html = requests.get(page_url).text
    bs= BeautifulSoup(html, 'html.parser')
    
    hrefs = bs.findAll('div', attrs={'class':'line thumbLi'})
    for href in hrefs :
        href_num = href.find('a')['href'][18:24]
        #print(href_num)
        url = 'http://www.koica.go.kr/bbs/koica_kr/147/{}/artclView.do'.format(href_num)
        
        html = requests.get(url).text

        bs= BeautifulSoup(html, 'html.parser')
                    
        dates = bs.findAll('div', attrs={'class': 'info'})
        for date in dates :
            d = str.strip(date.p.text)
            d = d[0:4]+"-"+d[5:7]+"-"+d[8:12]
            #print(d)
            result = [d]
            
        titles = bs.findAll('div', attrs={'class': 'heading'})
        for title in titles :
            t = str.strip(title.strong.text)
            #print(t)
            result.append(t)
            
        #print(url)
        result.append(url)
        
        print(result)
        wr.writerow(result)
        print(href_num,"처리 완료")
        
#'''

f.close()
print("*********************** 종료 ***********************")
```
