```python
# 기본 워드클라우드 생성하기
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
from wordcloud import WordCloud
text = open('alice.txt').read()
wordcloud = WordCloud().generate(text)
wordcloud.words_
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show
```

```python
# 특정 단어 제거하고 특정 이미지에 워드클라우드 입히기(said)
import numpy as np
from PIL import Image
from wordcloud import STOPWORDS
text = open('alice.txt').read()
alice_mask = np.array(Image.open("alice_mask.png"))
stopwords = set(STOPWORDS)
stopwords.add("said")
alice_mask
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()

wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask, stopwords=stopwords)
wc = wc.generate(text)
wc.words_
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# 회색으로 변형하기
import random
def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3), interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import nltk
import pillow
from wordcloud import WordCloud

from konlpy.corpus import kolaw
from konlpy.tag import Okt; okt = Okt()
ko_con_text = kolaw.open('constitution.txt').read()
ko_con_text
tokens_ko = okt.nouns(ko_con_text)
tokens_ko
stop_words = ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안']
token_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
token_ko
ko = nltk.Text(token_ko, name='대한민국 헌법')
ko.vocab().most_common(50)

import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="C:/windows/Fonts/H2GTRE.ttf").get_name()
rc('font', family=font_name)
get_ipython().run_line_magic('matplotlib', 'inline')
data = ko.vocab().most_common(500)
tmp_data = dict(data)
wordcloud = WordCloud(font_path='C:/windows/Fonts/H2GTRE.ttf', relative_scaling = 0.2, background_color='white', min_font_size=1, max_font_size=40).generate_from_frequencies(tmp_data)
plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()

```
