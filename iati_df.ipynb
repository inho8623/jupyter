```python
# 약정일을 날짜 타입으로 변경
iati['약정일'] = pd.to_datetime(iati['약정일'], format='%Y-%m-%d')
iati.dtypes
```

```python
# 약정일에서 약정연도 추출
iati['약정연도'] = pd.DatetimeIndex(iati['약정일']).year
iati.head()
```

```python
# 약정연도 데이터타입 확인
iati.dtypes
```

```python
# 기관별 약정연도별 피벗테이블
table_year = pd.pivot_table(iati, index='기관명', columns='약정연도', values='지출액(원화)', aggfunc='sum')
table_year
```

```python
# 엑셀로 저장
table_year.to_excel('pivot_sample.xlsx')
```

```python
koica = iati['기관명'] == '한국국제협력단(KOICA)'
koica
```

```python
# 기관명이 koica인 것만 필터링
is_koica = iati['기관명'] == '한국국제협력단(KOICA)'
koica = iati[is_koica]
koica
```

```python
# 약정연도 2016년 이후만 필터링
is_year = iati['약정연도'] >= 2016
subset_iati = iati[is_koica & is_year]
subset_iati
```

```python
# 피벗테이블 만들기
table_year2 = pd.pivot_table(subset_iati, index='기관명', columns='약정연도', values='지출액(원화)', aggfunc='sum')
table_year2
```

```python
# (숙제) 코이카랑 수은만 필터링
is_edcf = iati['기관명'] == '한국수출입은행(EDCF)'
subset_iati2 = iati[(is_koica | is_edcf) & is_year]
subset_iati2
```

```python
# (숙제?) 코이카랑 수은의 2016년 이후 데이터로 피벗 테이블 만들기
table_year3 = pd.pivot_table(subset_iati2, index='기관명', columns='약정연도', values='지출액(원화)', aggfunc='sum')
table_year3
```
