```python
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud
sdg_coloring = np.array(Image.open("sdg.png"))

from wordcloud import ImageColorGenerator
image_colors = ImageColorGenerator(sdg_coloring)

from wordcloud import STOPWORDS

text = open('sdg.txt', 'rt', encoding='UTF8').read()
stopwords = set(STOPWORDS)
stopwords.add("including")
wc = WordCloud(background_color="white", max_words=2000, mask=sdg_coloring, stopwords=stopwords, max_font_size=40, random_state=42)

wc.generate(text)

plt.figure(figsize=(12,12))
plt.imshow(sdg_coloring, cmap=plt.cm.gray, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()

plt.draw()
fig = plt.gcf()
fig.savefig('Kloud.png', dpi=fig.dpi)
```

```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Okt; okt = Okt()
```

```python
ko_con_text = kolaw.open('constitution.txt').read()
ko_con_text
```

```python
tokens_ko = okt.nouns(ko_con_text)
tokens_ko
```

```python
stop_words = ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안']

token_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
token_ko
```

```python
ko = nltk.Text(token_ko, name='대한민국 헌법')
ko.vocab().most_common(50)
```

```python
import matplotlib.pyplot as plt
import platform

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="C:/windows/Fonts/H2GTRE.ttf").get_name()
rc('font', family=font_name)

%matplotlib inline
```

```python
data = ko.vocab().most_common(500)
tmp_data = dict(data)

wordcloud = WordCloud(font_path='C:/windows/Fonts/H2GTRE.ttf', relative_scaling = 0.2, background_color='white', min_font_size=1, max_font_size=40).generate_from_frequencies(tmp_data)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
